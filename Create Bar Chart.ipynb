```python
#Add dependancies
import matplotlib.pyplot as plt
```

```python
#Set the x axis to a list of strings for each month
x_axis = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

#Set the y-axis to a list of floats as the tota fare in US Dollars accumualated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
#Create bar graph
plt.bar(x_axis, y_axis,color="green",label="Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot horizontally by switching x and y axis
plt.barh(x_axis, y_axis)
```

```python
# Create the plot horizontally by keeping axis
plt.barh(y_axis,x_axis)
```

```python
# Create the plot.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
#Skill Drill
#Using the Matplotlib MATLAB plotting approach, make the following changes:
#Change the bar colors to magenta.
#Add a legend for the city of Boston.
#Add a title and axes labels.
#Invert the y-axis so that January is at the top.
# Create the plot horizontally  
plt.barh(x_axis, y_axis,color = "magenta" , alpha=0.9, label="Boston")
plt.xlabel("Date")
plt.ylabel("Dollar($)")
plt.title ("PyBer Fare by Month") 
plt.gca().invert_yaxis()
```

```python
#Now use the object oriented programing to create bar chart
# Use ax.plt to create bar chart
fig  = plt.figure()
ax = fig.add_subplot()
ax.bar(x_axis,y_axis)
```

```python
##repeate the bar graph using bar color to black with a legend for the city of New York using the object-oriented interface method
fig,ax = plt.subplots()
ax.bar(x_axis,y_axis, color="black", label = "New York")
```

```python
#to create horizonal bar char using object oriented programming use ax.barh()
fig,ax = plt.subplots()
ax.barh(x_axis,y_axis)
```

```python
# horizontal bargraph using oo programming with axis data switched
fig,ax = plt.subplots()
ax.barh(y_axis,x_axis)
```

```python
#Using the object-oriented approach, make the following changes:
#Change the bar color to cyan.
#Add a legend for the city of Chicago.
#Add a title and axes labels.
#Switch the axis so that the Fare($) data is on the x-axis.
#Invert the y-axis data.
fix,ax = plt.subplots()
ax.barh (x_axis,y_axis, color ="c", label = "Chicago")
ax.invert_yaxis()
ax.set_xlabel("Dollar($)")
ax.set_ylabel("Date")
ax.legend()
ax.set_title("PyBer Fare by Month")
```
