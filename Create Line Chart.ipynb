```python
#import dependancies
import matplotlib.pyplot as plt
```

```python
#Set the x axis to a list of strings for each month
x_axis = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

#Set the y-axis to a list of floats as the tota fare in US Dollars accumualated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot()
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)  
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#Skill Drill
#Using the object-oriented approach, make the following changes to your line chart:
#Change the line color to green.
#Change the marker to a diamond.
#Change the line width to 2 points.
#Add a legend for the city of Boston.
#Add a title and axes labels.
#Add a y-axis limit.
#Add grid lines.
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis,color="green",marker="D",linewidth=2,label="Boston")  
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Month")
ax.set_ylabel("Fare($)")
ax.grid()
ax.set_ylim(0,45)
ax.legend()
```
