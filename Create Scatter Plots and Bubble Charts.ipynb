```python
#add dependaices
import matplotlib.pyplot as plt
```

```python
#Set the x axis to a list of strings for each month
x_axis = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

#Set the y-axis to a list of floats as the tota fare in US Dollars accumualated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#generate scattore plot using matplotlib plot . if we pass x and y data using plot method, it will create a line char. to covert it to scatter also pass a vairbale "o"
plt.plot(x_axis,y_axis,"o")
```

```python
#To create a scatter plot using plt .scatter
plt.scatter (x_axis,y_axis)
```

```python
#Using the Matplotlib MATLAB plotting approach, make the following changes to your scatter plot:
#Change the color of the markers to red.
#Add a legend for the city of Chicago.
#Add a title and axes labels.
#Switch the axis so that the Fare($) data is on the x-axis.
#Add a limit for the x-axis data.
#Invert the y-axis so that January is at the top.
plt.scatter (y_axis,x_axis, color="red",label ="Chicago")
plt.ylabel("Date")
plt.xlabel("Dollar($)")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
```

```python
#increase the size of bubles to make them more visible
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
#use python list comprehension technique to make the size bigger withing scatter method
plt.scatter(x_axis, y_axis, s = [i*3 for i in y_axis]) 
```

```python
#Create scattre plot using OO program mething
fig,ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#To create a bubble chart using the object-oriented interface approach, we'll add the s parameter just like we did when we used the MATLAB approach.

fig,ax = plt.subplots()
ax.scatter (x_axis,y_axis, s=y_axis)
```

```python
#Skill drill Using the object-oriented approach, make the following changes to your bubble chart:
#Change the color of the markers to sky blue.
#Change the size of the markers to 5 times each data point.
#Make the color 20% transparent.
#Add a black edge color to the circles.
#Make the linewidth of the circles 2 points.
#Add a legend for the city of Boston.
#Add a title.
#Switch the axis so that the Fare($) data is on the x-axis.
#Add a limit for the x-axis data.
#Invert the y-axis so that January is at the top
fix,ax = plt.subplots()
ax.scatter (y_axis, x_axis,facecolors = "skyblue" ,alpha = .8, edgecolors="black", s=[i*5 for i in y_axis],label="Boston",linewidth=2 )
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Dollar($)")
ax.set_ylabel("Date")
ax.set_xlim(0,45)
ax.invert_yaxis()
```
